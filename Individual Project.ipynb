```python
import pandas as pd
import numpy as np

#This section imports the Data into DataFrames who are able to be analyzed
Bs11L = pd.read_csv("11Bs Low.csv")
Bs11M = pd.read_csv("11Bs Med.csv")
Bs11H = pd.read_csv("11Bs High.csv")
Bs16S = pd.read_csv("16Bs Stable.csv")
Bs22M = pd.read_csv("22Bs Med.csv")
Bs43P1 = pd.read_csv("43Bs Proto1.csv")
Bs43P2 = pd.read_csv("43Bs Proto2.csv")
Bs64P1 = pd.read_csv("64Bs Proto1.csv")
Bs64P2 = pd.read_csv("64Bs Proto2.csv")
Bs86P1 = pd.read_csv("86Bs Proto1.csv")
Bs86P2 = pd.read_csv("86Bs Proto2.csv")
Bs172 = pd.read_csv("172Bs Proto1.csv")
Bs258 = pd.read_csv("258Bs Proto1.csv")

def Perf(Proto):
    """Given a Protocol's Dataset, returns the number of trials with perfect output
    """
    perf = np.zeros(5)
    for i in range(0,5):
        for j in range(0,15):
            #Analyzes data to find which transmissions had perfect output, and which had imperfect output
            #Does this by checking if each transmission is exactly equal to "Sample Data"
            #The imperfect output percent is checked in the creation of the final DataFrame by subtracting the Perfect output % by 100
            if Proto.iloc[j,i] == "Sample Data":
                perf[i] = perf[i] + 1
    return perf

def SOut(Proto):
    """Given a Protocol's Dataset, returns the number of trials with some output
    """
    SOut = np.zeros(5)
    for i in range(0,5):
        for j in range(0,15):
            #This section counts up the number of transmissions that had some amount of output
            #   does this by checking if each transmissions is not 'NaN'
            #The No Output counter is done in the creation of the final data
            #   frame by subtracting the Some output percent by 100
            if pd.isna(Proto.iloc[j,i]) != True:
                SOut[i] = SOut[i] + 1
    return SOut

def MFrame(Proto):
    """Given a Protocol's Dataset, returns the number of trials made up of multiple frames
    """
    MultipleFramesAr = np.zeros(5)
    for i in range(0,5):
        for j in range(0,15):
            if '\\' in str(Proto.iloc[j,i]):
                MultipleFramesAr[i] = MultipleFramesAr[i] + 1
    return MultipleFramesAr

def SFrame(Proto):
    """Given a Protocol's Dataset, returns the number of trials made up of single frames
    """
    SingleFrameAr = np.zeros(5)
    for i in range(0,5):
        for j in range(0,15):
            if ('\\' not in str(Proto.iloc[j,i])) & (pd.isna(Proto.iloc[j,i]) != True):
                SingleFrameAr[i] = SingleFrameAr[i] + 1
    return SingleFrameAr
            
#This section creates the final DataFrame by creating a series for each of the criteria,
#and calculating the percentages for each protocol
SomeOutput = pd.Series({'11B/s, Low Freq': ((SOut(Bs11L).sum()/75)*100), '11B/s, Med Freq': ((SOut(Bs11M).sum()/75)*100),
                        '11B/s, High Freq': ((SOut(Bs11H).sum()/75)*100), '16B/s, Stable': ((SOut(Bs16S).sum()/75)*100),
                        '22B/s, Med Freq': ((SOut(Bs22M).sum()/75)*100), '43B/S, Protocol 1': ((SOut(Bs43P1).sum()/75)*100),
                        '43B/s, Protocol 2': ((SOut(Bs43P2).sum()/75)*100), '64B/s, Protocol 1': ((SOut(Bs64P1).sum()/75)*100), 
                        '64B/s, Protocol 2': ((SOut(Bs64P2).sum()/75)*100), '86B/s, Protocol 1': ((SOut(Bs86P1).sum()/75)*100),
                        '86B/s, Protocol 2': ((SOut(Bs86P2).sum()/75)*100), '172B/s, Protocol 1': ((SOut(Bs172).sum()/75)*100),
                        '258B/s, Protocol 1': ((SOut(Bs258).sum()/75)*100)})
NoOutput = pd.Series({'11B/s, Low Freq': (-SomeOutput[0]+100), '11B/s, Med Freq': (-SomeOutput[1]+100),
                        '11B/s, High Freq': (-SomeOutput[2]+100), '16B/s, Stable': (-SomeOutput[3]+100),
                        '22B/s, Med Freq': (-SomeOutput[4]+100), '43B/S, Protocol 1': (-SomeOutput[5]+100),
                        '43B/s, Protocol 2': (-SomeOutput[6]+100), '64B/s, Protocol 1': (-SomeOutput[7]+100), 
                        '64B/s, Protocol 2': (-SomeOutput[8]+100), '86B/s, Protocol 1': (-SomeOutput[9]+100),
                        '86B/s, Protocol 2': (-SomeOutput[10]+100), '172B/s, Protocol 1': (-SomeOutput[11]+100),
                        '258B/s, Protocol 1': (-SomeOutput[12]+100)})

PerfectOutput = pd.Series({'11B/s, Low Freq': ((Perf(Bs11L).sum()/75)*100), '11B/s, Med Freq': ((Perf(Bs11M).sum()/75)*100),
                        '11B/s, High Freq': ((Perf(Bs11H).sum()/75)*100), '16B/s, Stable': ((Perf(Bs16S).sum()/75)*100),
                        '22B/s, Med Freq': ((Perf(Bs22M).sum()/75)*100), '43B/S, Protocol 1': ((Perf(Bs43P1).sum()/75)*100),
                        '43B/s, Protocol 2': ((Perf(Bs43P2).sum()/75)*100), '64B/s, Protocol 1': ((Perf(Bs64P1).sum()/75)*100), 
                        '64B/s, Protocol 2': ((Perf(Bs64P2).sum()/75)*100), '86B/s, Protocol 1': ((Perf(Bs86P1).sum()/75)*100),
                        '86B/s, Protocol 2': ((Perf(Bs86P2).sum()/75)*100), '172B/s, Protocol 1': ((Perf(Bs172).sum()/75)*100),
                        '258B/s, Protocol 1': ((Perf(Bs258).sum()/75)*100)})
ImperfectOutput = pd.Series({'11B/s, Low Freq': (-PerfectOutput[0]+100), '11B/s, Med Freq': (-PerfectOutput[1]+100),
                        '11B/s, High Freq': (-PerfectOutput[2]+100), '16B/s, Stable': (-PerfectOutput[3]+100),
                        '22B/s, Med Freq': (-PerfectOutput[4]+100), '43B/S, Protocol 1': (-PerfectOutput[5]+100),
                        '43B/s, Protocol 2': (-PerfectOutput[6]+100), '64B/s, Protocol 1': (-PerfectOutput[7]+100), 
                        '64B/s, Protocol 2': (-PerfectOutput[8]+100), '86B/s, Protocol 1': (-PerfectOutput[9]+100),
                        '86B/s, Protocol 2': (-PerfectOutput[10]+100), '172B/s, Protocol 1': (-PerfectOutput[11]+100),
                        '258B/s, Protocol 1': (-PerfectOutput[12]+100)})

SingleFrame = pd.Series({'11B/s, Low Freq': ((SFrame(Bs11L).sum()/75)*100), '11B/s, Med Freq': ((SFrame(Bs11M).sum()/75)*100),
                        '11B/s, High Freq': ((SFrame(Bs11H).sum()/75)*100), '16B/s, Stable': ((SFrame(Bs16S).sum()/75)*100),
                        '22B/s, Med Freq': ((SFrame(Bs22M).sum()/75)*100), '43B/S, Protocol 1': ((SFrame(Bs43P1).sum()/75)*100),
                        '43B/s, Protocol 2': ((SFrame(Bs43P2).sum()/75)*100), '64B/s, Protocol 1': ((SFrame(Bs64P1).sum()/75)*100), 
                        '64B/s, Protocol 2': ((SFrame(Bs64P2).sum()/75)*100), '86B/s, Protocol 1': ((SFrame(Bs86P1).sum()/75)*100),
                        '86B/s, Protocol 2': ((SFrame(Bs86P2).sum()/75)*100), '172B/s, Protocol 1': ((SFrame(Bs172).sum()/75)*100),
                        '258B/s, Protocol 1': ((SFrame(Bs258).sum()/75)*100)})
MultipleFrames = pd.Series({'11B/s, Low Freq': ((MFrame(Bs11L).sum()/75)*100), '11B/s, Med Freq': ((MFrame(Bs11M).sum()/75)*100),
                        '11B/s, High Freq': ((MFrame(Bs11H).sum()/75)*100), '16B/s, Stable': ((MFrame(Bs16S).sum()/75)*100),
                        '22B/s, Med Freq': ((MFrame(Bs22M).sum()/75)*100), '43B/S, Protocol 1': ((MFrame(Bs43P1).sum()/75)*100),
                        '43B/s, Protocol 2': ((MFrame(Bs43P2).sum()/75)*100), '64B/s, Protocol 1': ((MFrame(Bs64P1).sum()/75)*100), 
                        '64B/s, Protocol 2': ((MFrame(Bs64P2).sum()/75)*100), '86B/s, Protocol 1': ((MFrame(Bs86P1).sum()/75)*100),                        
                        '86B/s, Protocol 2': ((MFrame(Bs86P2).sum()/75)*100), '172B/s, Protocol 1': ((MFrame(Bs172).sum()/75)*100),
                        '258B/s, Protocol 1': ((MFrame(Bs258).sum()/75)*100)})

Final = pd.DataFrame({'% Some Output':SomeOutput.round(2), '% No Output':NoOutput.round(2),
                      '% Perfect Output':PerfectOutput.round(2), '% Imperfect Output':ImperfectOutput.round(2),
                      '% Single Frame':SingleFrame.round(2), '% Multiple Frames':MultipleFrames.round(2)})

FinalRanking = pd.DataFrame({'% Some Output':SomeOutput.rank(method='min',ascending=False).astype(int),
                             '% No Output':NoOutput.rank(method='min').astype(int),
                             '% Perfect Output':PerfectOutput.rank(method='min',ascending=False).astype(int),
                             '% Imperfect Output':ImperfectOutput.rank(method='min').astype(int),
                             '% Single Frame':SingleFrame.rank(method='min',ascending=False).astype(int),
                             '% Multiple Frames':MultipleFrames.rank(method='min',ascending=True).astype(int)})

#This section creates a table that displays the ratio of the percent
#of Perfect trials to the percent of trials with some output
Ratios  = np.zeros(13)
for  i in range(0,13):
    Ratios[i] = Final.iloc[i,2] / Final.iloc[i,0]

RatioSomePerfBasic = pd.Series({'11B/s, Low Freq': Ratios[0], '11B/s, Med Freq': Ratios[1],
                        '11B/s, High Freq': Ratios[2], '16B/s, Stable': Ratios[3],
                        '22B/s, Med Freq': Ratios[4], '43B/S, Protocol 1': Ratios[5],
                        '43B/s, Protocol 2': Ratios[6], '64B/s, Protocol 1': Ratios[7], 
                        '64B/s, Protocol 2': Ratios[8], '86B/s, Protocol 1': Ratios[9],
                        '86B/s, Protocol 2': Ratios[10], '172B/s, Protocol 1': Ratios[11],
                        '258B/s, Protocol 1': Ratios[12]})
RatioSomePerfFinal = pd.DataFrame({'Ratio of  “% Perfect Output” to “% Some Output”': RatioSomePerfBasic.round(3),
                                   'Rank': RatioSomePerfBasic.rank(method='max',ascending=False).round(1).astype(int)})   
```
